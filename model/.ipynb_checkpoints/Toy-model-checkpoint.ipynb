```python
import numpy as np
import pandas as pd
import torch
from collections import namedtuple, deque
from torch import nn
import torch
from torch import optim
import torch.nn.functional as F
import random
from collections import namedtuple, deque
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
import operator
import copy
```

```python
KIND_CARS = 8
distances = np.ones(KIND_CARS)
penalties = np.ones(KIND_CARS)
distances[0:(int(KIND_CARS/2))] = 2
distances[(int(KIND_CARS/2)):] = 3
penalties[0:(int(KIND_CARS/2))] = -5
penalties[(int(KIND_CARS/2)):] = -3

INPUT_SEQUENCE_LENGTH = 30
INPUT_WINDOW = 3
OUTPUT_SEQUENCE_LENGTH = 5
NUM_LINES = 2
CAPACITY_LINES = 3
```

```python
def calculate_window(sequence, index, offset):
    low = int(np.max([0, index - offset]))
    high = int(index+1)
    return torch.tensor(sequence)[low: high]

def judge_partial_sequence(sequence):
    index = len(sequence)-1
    car = sequence[index]
    
    window = calculate_window(sequence, index, distances[car])
    failures = (window == car).sum().item() - 1
    reward = failures * penalties[car]
    
    return reward
```

```python
class environment():
    def __init__(self,):

        self.length = INPUT_SEQUENCE_LENGTH
        self.input_sequence = torch.randint(0, KIND_CARS, (self.length,))
        self.buffer = torch.ones((NUM_LINES, CAPACITY_LINES)).long() * (-1)
        self.output_sequence = deque(maxlen=OUTPUT_SEQUENCE_LENGTH)
        for i in range(OUTPUT_SEQUENCE_LENGTH):
            self.output_sequence.append(-1)
        
        self.done = False
        
    
    def step(self, action):
        if self.done:
            print("Already done, step won't have an effect")
            return None, None, None
        car = self.input_sequence[0]
        self.input_sequence = self.input_sequence[1:]
        out = self.buffer[action][-1].item()
        line = self.buffer[action][:-1]
        self.buffer[action] = torch.cat((torch.tensor([car]), line))
        
        self.done = self.input_sequence.size()[0] == 0
        
        
        reward = 0
        if out != -1:
            pos = len(self.output_sequence)
            self.output_sequence.append(out)
            reward = judge_partial_sequence(self.output_sequence)
        
        next_state = self.get_state()
        return reward, next_state, self.done
    
    def reset(self, given_sequence = []):
        if len(given_sequence) != 0:
            self.input_sequence = given_sequence
        else:
            self.input_sequence = torch.randint(0, KIND_CARS, (self.length,))
        self.buffer = torch.ones((NUM_LINES, CAPACITY_LINES)).long() * (-1)
        
        self.output_sequence = deque(maxlen=OUTPUT_SEQUENCE_LENGTH)
        for i in range(OUTPUT_SEQUENCE_LENGTH):
            self.output_sequence.append(-1)
        
        self.done = False
        state = self.get_state()
        return state
    
    def get_state(self):
        inp = self.input_sequence[0:(INPUT_WINDOW)].tolist()
        while len(inp) < INPUT_WINDOW:
            inp.append(-1)
        return inp +  self.buffer.resize(CAPACITY_LINES*NUM_LINES).tolist() + list(self.output_sequence)  
    
    def show(self):
        print("input: ", self.input_sequence)
        print("buffer: ", self.buffer)
        print("output: ", self.output_sequence)
```

```python
# Constants
BUFFER_SIZE = int(1e4)  # replay buffer size
BATCH_SIZE = 64         # minibatch size
GAMMA = 1          # discount factor
TAU = 0.001#1e-3              # for soft update of target parameters
LR = 5e-4               # learning rate 
UPDATE_EVERY = 4        # how often to update the network

best_score = - float("inf")
```

```python
# define neural network
class Network(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear((INPUT_WINDOW + OUTPUT_SEQUENCE_LENGTH+NUM_LINES*CAPACITY_LINES)*(KIND_CARS+1), 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 64)
        self.fc4 = nn.Linear(64, NUM_LINES)
    def forward(self, x):
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)
        x = F.relu(x)
        x = self.fc4(x)
        return x
```

```python
#Replay Buffer used for Experience Replay
class ReplayBuffer:
    
    def __init__(self, buffer_size, batch_size, seed):
        self.batch_size = batch_size
        self.memory = deque(maxlen = buffer_size)
        self.seed = random.seed(seed)
    
    def add(self, state, action, reward, next_state, done):
        if done:
            done_value = 1
        else:
            done_value = 0
        self.memory.append([state, action, reward, next_state, done_value])
    
    def sample(self):
        samples = (random.sample(self.memory, self.batch_size))
        states = []
        actions = []
        rewards = []
        next_states = []
        dones = []
        
        for sample in samples:
            state, action, reward, next_state, done = sample
            
            states.append(state)
            actions.append(action)
            rewards.append(reward)
            next_states.append(next_state)
            dones.append(done)

        states = torch.tensor(states).float()
        actions = torch.LongTensor(actions)
        rewards = torch.tensor(rewards).float()
        next_states = torch.tensor(next_states).float()
        dones = torch.tensor(dones).float()
        
        return [states,actions,rewards,next_states,dones]
        
    
    def __len__(self):
        return len(self.memory)
```

```python
# Agent
class Agent():
    def __init__(self, seed):
        self.qnetwork_local = Network()
        self.qnetwork_target = Network()
        self.seed = random.seed(seed)
        
        self.optimizer = optim.Adam(self.qnetwork_local.parameters(), lr=LR)
        self.memory = ReplayBuffer(BUFFER_SIZE, BATCH_SIZE, seed)
        
        self.t_step = 0
        
    def step(self, state, action, reward, next_state, done):
        self.memory.add(state, action, reward, next_state, done)
        
        self.t_step = (self.t_step + 1) % UPDATE_EVERY
        if self.t_step == 0:
            if (len(self.memory)) > BATCH_SIZE:
                samples = self.memory.sample()
                self.learn(samples, GAMMA)
    
    def act(self, state, eps = 0):
        state = torch.tensor(state).float()
        with torch.no_grad():
            action_values = self.qnetwork_local(state)
        
        if random.random()>eps:
            return np.argmax(action_values.numpy())
        else:
            return random.choice(range(len(action_values.numpy())))
    
    def learn(self, samples, gamma):
        states, actions, rewards, next_states, dones = samples
        
        q_values_next_states = self.qnetwork_target.forward(next_states).max(dim=1)[0]#.unsqueeze(1)
        targets = rewards + (gamma * (q_values_next_states) *  (1 - dones))
        q_values = self.qnetwork_local.forward(states)


        actions = actions.view(actions.size()[0], 1)
        predictions = torch.gather(q_values, 1, actions).view(actions.size()[0])
         
        loss = F.mse_loss(predictions, targets)
        
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
        
        for target_param, local_param in zip(self.qnetwork_target.parameters(), self.qnetwork_local.parameters()):
            target_param.data.copy_(TAU*local_param.data + (1.0-TAU)*target_param.data)
#         self.soft_update(self.qnetwork_local, self.qnetwork_target, TAU)      
 
    def soft_update(self, local_model, target_model, tau):
        for target_param, local_param in zip(target_model.parameters(), local_model.parameters()):
            target_param.data.copy_(tau*local_param.data + (1.0-tau)*target_param.data)
```

```python
def train(n_episodes=8000, max_t=10000, agent = None, checkpoint_name = 'Unnamed.pth',  eps_start=1.0, eps_end=0.0001, eps_decay=0.999):
    global best_score
    if agent == None:
        agent = Agent(0)
    # status variables and plots
    scores = []  # list containing scores from each episode
    means = []
    scores_window = deque(maxlen=100)  # last 100 scores
    eps = eps_start                    # initialize epsilon
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    for i_episode in range(1, n_episodes+1):
        state = env.reset()
        score = 0
        for t in range(max_t):
            action = agent.act(state, eps)
            reward, next_state, done = env.step(action) # send the action to the environment and observe
            agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward*np.power(GAMMA, t)
            if done:
                break
        scores_window.append(score)
        scores.append(score)
        eps = max(eps_end, eps_decay*eps)
        
        means.append(np.mean(scores_window))
        
        if i_episode % 100 == 0:
            
            score = np.mean(scores_window)
            
            # if current score is better, save the network weights and update best seen score
            if score > best_score: 
                best_score = score 
                torch.save(agent.qnetwork_local.state_dict(), checkpoint_name + '.pth')
            clear_output(wait=True)
            plt.scatter(range(len(scores)), scores, label = 'Scores', color = 'c', alpha = 0.8, s=1)
            plt.plot(np.arange(len(means)),means, label = 'Mean', color = 'r')
            plt.ylabel('Score')
            plt.xlabel('Episode #')
            plt.show()
            print('\rEpisode {}\tAverage Score: {:.2f}\tBest Score: {:.2f}'.format(i_episode, score, best_score))

            f = open(checkpoint_name + ".scores", "a")
            for score in scores_window:
                f.write(str(score) + '\n')
            f.close()
    return agent, scores


```

```python
class heuristic():
    def __init__(self):
        self.seen = []
        
    def act(self, state):
        out0 = state[INPUT_WINDOW + CAPACITY_LINES -1]
        out1 = state[INPUT_WINDOW + 2*CAPACITY_LINES -1]
        
        for car in self.seen[::-1]:
            if out1 == car:
                self.seen.append(out0)
                return 0
            if out0 == car: 
                self.seen.append(out1)
                return 1
        self.seen.append(out0)
        return 0

    def reset(self):
        self.seen = []
```

```python
class Search_node():
    def __init__(self,value, visited = [], done = False):
        self.visited = visited
        self.value = value
        self.done = done

    def explore(self):
        res = []
        for i in range(NUM_LINES):
            child = copy.deepcopy(self.visited)
            child.append(i)
            res.append(child)
        return res
    
    def show(self):
        print(self.value, self.visited)
        
    def terminal(self):
        return self.done

def optimal(test_sequence):
    result = None
    root = Search_node(0)
    open_list = [root]
    while len(open_list) != 0:
        actual = open_list.pop()
        if actual.terminal():
#             print("Best result found!")
#             actual.show()
            result = actual
            break
        children = actual.explore()
        for child in children:
            value, done = follow_decision(test_sequence, child)
            open_list.append(Search_node(value, child, done))
        open_list.sort(key=operator.attrgetter('value'))
    
    return result.value
```

```python
def one_hot(states):
    length = len (states)
    occ = np.ones(length)
    occ[states == -1] = 0
    res = np.zeros((length, KIND_CARS))
    for i,state in enumerate(states):
        if occ[i]:
            res[i,state] = 1
        
    return occ, res

def linearize(states):
    length = len(states)
    occ, states = one_hot(states)
    states = states.reshape(length*(KIND_CARS))
    return np.concatenate([occ, states])
```

```python
def oh_train(n_episodes=8000, max_t=10000, agent = None, checkpoint_name = 'Unnamed.pth',  eps_start=1.0, eps_end=0.0001, eps_decay=0.999):
    global best_score
    if agent == None:
        agent = Agent(0)
    # status variables and plots
    scores = []  # list containing scores from each episode
    means = []
    scores_window = deque(maxlen=100)  # last 100 scores
    eps = eps_start                    # initialize epsilon
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    for i_episode in range(1, n_episodes+1):
        state = env.reset()
        state = linearize(state)
        score = 0
        for t in range(max_t):
            action = agent.act(state, eps)
            reward, next_state, done = env.step(action) # send the action to the environment and observe
            next_state = linearize(next_state)
            agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward*np.power(GAMMA, t)
            if done:
                break
        scores_window.append(score)
        scores.append(score)
        eps = max(eps_end, eps_decay*eps)
        
        means.append(np.mean(scores_window))
        
        if i_episode % 100 == 0:
            
            score = np.mean(scores_window)
            
            # if current score is better, save the network weights and update best seen score
            if score > best_score: 
                best_score = score 
                torch.save(agent.qnetwork_local.state_dict(), checkpoint_name + '.pth')
            clear_output(wait=True)
            plt.scatter(range(len(scores)), scores, label = 'Scores', color = 'c', alpha = 0.8, s=1)
            plt.plot(np.arange(len(means)),means, label = 'Mean', color = 'r')
            plt.ylabel('Score')
            plt.xlabel('Episode #')
            plt.show()
            print('\rEpisode {}\tAverage Score: {:.2f}\tBest Score: {:.2f}'.format(i_episode, score, best_score))

            f = open(checkpoint_name + ".scores", "a")
            for score in scores_window:
                f.write(str(score) + '\n')
            f.close()
    return agent, scores



```
